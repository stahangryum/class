```r
x <- seq(0, 5, 0.1)
y1 <- x
y2 <- 2*sqrt(x+1)-2

par(mfrow=c(1,2))
plot(y1 ~ x, type ='l', xlim = c(-1, 5), ylim = c(-1, 5), ylab = 'y', main = 'x^(1/2) transform', lty='dotted')
par(new = T)
plot(y2 ~ x, type ='l', xlim = c(-1, 5), ylim = c(-1, 5), ylab = '', xlab = '')

y1 <- y2
y2 <- 4*((x+1)^(1/4) - 1)
plot(y1 ~ x, type = 'l', xlim = c(-1, 5), ylim = c(-1, 5), ylab = 'y', main = 'x^(1/4) transform', lty='dotted')
par(new = T)
plot(y2 ~ x, type = 'l', xlim = c(-1, 5), ylim = c(-1, 5), ylab = '', xlab ='')
```

# 5.2 재표현의 목적

`1` 여러 자료 묶음(그룹)의 산포가 상이하여 그룹간 비교가 어려운 경우

`2` 변수간 관계가 비 선형적인 경우

## 여러 자료 묶음의 산포가 상이하여 그룹간 비교가 어려운 경우

```r
x1 = rgamma(100, 4); g1 = rep('A', 100)
x2 = rgamma(100, 5); g2 = rep('B', 100)
x3 = rgamma(100, 9); g3 = rep('C', 100)

claims = c(x1, x2, x3)
sqrt_claims = sqrt(claims)
log_claims = log(claims)
haha_claims = claims ** (1/2^40)

group = factor(c(g1, g2, g3))

par(mfrow = c(2,2))
boxplot(claims ~ group, ylab = 'raw scale')
boxplot(sqrt_claims ~ group, ylab = 'sqrt scale')
boxplot(log_claims ~ group, ylab = 'log scale')
boxplot(haha_claims ~ group, ylab = 'haha')
```

## 변수간 관계가 비선형적인 경우

```r
par(mfrow=c(1,2))
x <- rnorm(100, 5, 1.5)
y1 <- 0.1*x^2 + 0.2*rnorm(100, 0, 1)
plot(y1 ~ x, main = 'a*x^2 + b')
y2 <- 10*sqrt(x) + 0.2*rnorm(100, 0, 1)
plot(y2 ~ x, main = 'a*sqrt(x) + b')
```

```r
# y변수를 변환하는 경우

par(mfrow=c(1,2)) 

plot(sqrt(y1) ~ x, col = 'red')
par(new=T)
plot(y1 ~ x, col = 'blue', ylab = '', xlab = '')
legend('topleft', c('변환 이전', '변환 이후'), fill = c('blue', 'red'))

plot(y2^2 ~ x, col = 'red')
par(new=T)
plot(y2 ~ x, col = 'blue', ylab = '', xlab ='')
legend('topleft', c('변환 이전', '변환 이후'), fill = c('blue', 'red'))
```
