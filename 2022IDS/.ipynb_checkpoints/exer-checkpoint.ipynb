## 02

```r
library(DAAG)
data(greatLakes)
ds <- data.frame(year = 1918:2009, greatLakes)
head(ds)
```

```r
plot(ds$Erie, type = 'l')
```

```r
plot(ds$year, ds$michHuron, 
     type = 'b', 
     col = 'red',
     main = 'michHuron호 수위',
     xlab = '연도',
     ylab = '수위')
```

```r
plot(ds$year, ds$Erie,
     type = 'b',
     col = 'red',
     ylim = c(173, 177.5))

lines(ds$year, ds$michHuron,
      col = 'blue',
      type = 'b')

lines(ds$year, ds$StClair,
      col = 'purple',
      type = 'b')
```

## 03

```r
library(DAAG)
```

```r
boxplot(cfseal$weight)
```

```r
boxplot.stats(cfseal$heart)$out
```

```r
group <- ifelse(cfseal$age < mean(cfseal$age), 'young', 'old')
```

```r
boxplot(cfseal$weight ~ group,
        col = c('orange', 'green'))
```

```r
Q1 <- boxplot.stats(cfseal$weight)$stats[c(2)]
med <- boxplot.stats(cfseal$weight)$stats[c(3)]
Q3 <- boxplot.stats(cfseal$weight)$stats[c(4)]
```

```r
group <- ifelse(cfseal$weight < Q1, 'low',
                ifelse((Q1 < cfseal$weight) & (cfseal$weight < Q3),
                       'middle', 'high'))
```

```r
boxplot(cfseal$stomach ~ group)
```
