```r
if (!require(rvest)) install.packages('rvest')
library(rvest)

if (!require(tidyverse)) install.packages('tidyverse')
library(tidyverse)

target_url <- 'https://ko.wikipedia.org/wiki/%EC%A1%B0%EC%84%A0_%EA%B5%AD%EC%99%95'
html <- read_html(target_url, encoding = 'UTF-8')

order = 1

unclean_tables <- html %>%
    html_nodes('table[class=wikitable]') %>%
    html_nodes('tbody') %>%
    html_nodes('tr') %>%    
    html_text()
```

```r
true_data_index <- unclean_tables %>% str_detect('^제')
true_table <- unclean_tables[true_data_index][order] %>% # 한글자씩 쪼갠다.
    strsplit(split = '') %>%
    unlist() %>%
    str_replace('(![가-힣0-9])&(\r\n)', 'N') # gsub vs str_replace ??

true_table
```
