```r
if (!require(tidyverse)) install.packages('tidyverse')
library(tidyverse)


# 애초에 파일을 부분적으로 읽는 방법은 없나?

pi <- substr(readLines('pi_hexadecimal.txt'), 3, 10^7 + 2)

pi_splited <- 
    pi %>%
    strsplit(split = NULL) %>%
    unlist()

pi_splited

paste(pi_splited[(length(pi_splited)-5) : length(pi_splited)], collapse = '')

hex_box <- array() # 16진수 중 중복되지 않은 원소를 발견하면 기록한다.
hex_count <- 0 # 인덱싱을 편하게 하기 위한 도구

complete_hex_index <- array() # 16진수를 모두 모을 때마다 당시의 인덱스를 기록한다.
complete_count <- 0 # 인덱싱을 편하게 하기 위한 도구

for (i in 1:length(pi_splited)){
    if (!pi_splited[i] %in% hex_box){ # 중복된 원소가 존재하지 않으면
        hex_count <- hex_count + 1
        hex_box[hex_count] <- pi_splited[i]
        
        if (hex_count == 16){ # 중복되지 않은 16진수를 모두 찾으면  
            hex_box <- array() # 초기화
            hex_count <- 0 # 카운트 초기화
            
            complete_count <- complete_count + 1
            complete_hex_index[complete_count] <- i # 최종 원소의 인덱스 기록
        }
    }
}    

index_diff <- diff(c(0, complete_hex_index))

table(index_diff)

calc_coupon_expect <- function(n){
    E <- n*sum(1/(1:n))
    return (E)
}

p_n <- function(n){
    k = 16
    p = (1/(k)^(n-1)) * sum((-1)^(0:(k-1)) * choose(k-1, 0:(k-1)) * (k-1-(0:(k-1)))**(n-1))
    return(p)
}

get_graph <- function(n){ # 자리수를 입력하면 그래프를 반환하는 함수를 정의함
    index_diff_until_n <- diff(c(0, complete_hex_index[complete_hex_index <= n]))
    hist(index_diff_until_n,
         prob = TRUE,
         breaks = seq(1,max(index_diff_until_n), 1),
         main = paste(n, ' 자리\n 평균=', round(mean(index_diff_until_n), 2)),
         xlab = 'length')
    
    theorical_result <- array()
    for (i in 1:max(index_diff_until_n)) theorical_result[i] <- p_n(i)
    
    lines(theorical_result, col = 'red')
}

par(mfrow = c(2,2))
get_graph(10**4)
get_graph(10**5)
get_graph(10**6)
get_graph(10**7)

cat("\n1) '10^7+2' 개 문자를 읽는다. 맨 앞의 문자 두 개를 제외한 나머지 10^7 개 문자를 저장한다. 문자를 하나씩 분리한다. 마지막 문자 6개는",
    paste(pi_splited[(length(pi_splited)-5) : length(pi_splited)], collapse = ''), '이다.', sep = '')

cat('\n2) 16진수 16개 숫자를 모두 모을 때까지 얼마나 많은 숫자가 필요할까? 처음은 ', complete_hex_index[1],', 그 다음은 ', 
    complete_hex_index[2],', ', complete_hex_index[3], '이다. 10^7개 중에 총 ', length(complete_hex_index), '개가 있다.', sep = '')

cat('\n3) 가장 짧은 경우는 ', min(index_diff), '개이며 총 ', table(index_diff)[1], '번 나왔다. 가장 긴 경우는 ',
    max(index_diff), '개이며 ', table(index_diff)[length(table(index_diff))], '번 나왔다.', sep = '')

cat('\n4) 이론적인 기댓값은 여기를 참조하라. 기댓값을 소숫점이하 두자리까지 구하면 ', round(calc_coupon_expect(16), 2), '이다.', sep = '')
```
