- 벡터 안에서 연속된 값 찾기

```r
set.seed(1)
poisson <- rpois(150, 0.7)
poisson
```

```r
rle(poisson)
```

```r
which(rle(poisson)[[1]] == max(rle(poisson)[[1]]))
```

```r
rle(poisson)[[2]][26]
```

```r
n1 <- 25
n2 <- 30

y <- c(rep(1, n1), rep(0, n2))
y
```

```r
set.seed(1)
sample(y)
```

```r
len <- numeric(10000)
for (i in 1:10000) len[i] = length(rle(sample(y))[[2]])
```

```r
hist(len)
```

```r
set.seed(1)
A = sample(6, 1000000, replace = TRUE)
```

```r
A
```

```r
which(A == 6)[1]
```

```r
which(A == 6)[2] - which(A == 6)[1]
```

```r
c(0, which(A == 6)) %>% diff() %>% mean()
```

```r
c(0, which(A == 6)) %>% diff() %>% table() %>% prop.table() %>% barplot()
```
