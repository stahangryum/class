```r
n1 <- 300
x1 <-  runif(n1, min = 0, max = 100)
```

```r
y1 <- 25 + 0.4*x1 + rnorm(n1, mean=0, sd=5)
```

```r
plot(x1, y1, col = 'red', xlab = 'Explanatory variable', ylab = 'Response variable')
abline(h=40)
abline(v=20)
abline(lm(y1~x1))
```

```r
plot(x1, y1, col = 'red', xlab = 'Explanatory variable', ylab = 'Response variable')
#abline(h=40)
#abline(v=20)
abline(lm(y1~x1))

set.seed(2)
n2 <- 200
x2 <- runif(n2, min = 0, max = 100)
y2 <- 15 + 0.5*x2 + rnorm(n2, mean = 0, sd = 5)

points(x2, y2, col = 'blue', pch = 16)
abline(lm(y2~x2))
```

```r
plot(c(x1, x2), c(y1, y2), xlab = 'Explanatory variable', ylab = 'Response variable', type = 'n')

points(x1, y1, col = 'red')
points(x2, y2, col = 'blue', pch = 16)

abline(lm(y1~x1))
abline(lm(y2~x2))
```

```r
range(c(x1,x2))
```

```r
plot(c(x1, x2), c(y1, y2), xlim = range(c(x1,x2)), ylim = range(c(y1,y2)), xlab = 'Explanatory variable', ylab = 'Response variable')

abline(lm(y1~x1))
abline(lm(y2~x2))
```

```r
plot(0, 0, xlim = c(0,15), ylim = c(1,10), type ='n', xlab ='', ylab ='')

x <- seq(1, 13, 2) 
s <- -7
f <- -1

for (y in seq(9,3,-2)) {
  # x : 1 3 5 7 9 11 13
  # y : 9, 7, 5, 3
  s <- s+7 # S : 0, 7, 14, 21
  f <- f+7 # F : 6, 13, 20, 27
  y2 <- rep(y, length(x)) # 9
  print(y2)
  points(x, y2, pch=s:f, cex=1.5)   # cex 옵션은 점 크기
 
  text(x, y-1, as.character(s:f), cex=2) 
}
```

```r
sleep <- read.table("http://www.bio.ic.ac.uk/research/mjcraw/therbook/data/sleep.txt", header=TRUE)
```

```r
str(sleep)
```

```r
head(sleep)
```

```r
plot(sleep$Days, sleep$Reaction, type = 'n', xlab = 'Days', ylab = 'Reaction')
s <- as.numeric(factor(sleep$Subject))
for (k in 1:max(s)){
    x <- sleep$Days[s == k]
    y <- sleep$Reaction[s == k]
    points(sleep$Days[s == k], sleep$Reaction[s==k], pch = sym[k], bg = bcol[k], col = 1)
    lines(x, y, type = 'b')
}
```

```r
length(unique(s))
```

```r
plot(sleep$Days, sleep$Reaction, type = 'n', xlab = 'Days', ylab = 'Reaction')
sym <- rep(c(21, 22, 24), c(7, 7, 4))   # 점 기호는 3종류 
bcol <- c(2:8, 2:8, 2:5)   # 피험자별 배경 색 종류

for (k in 1:max(s)){
    points(sleep$Days[s == k], sleep$Reaction[s==k], pch = sym[k], bg = bcol[k], col = 1)
}
```

```r
plot(0:10, 0:10, xlab = '', ylab = '', xaxt = 'n', yaxt = 'n', type = 'n')
rect(6,6,9,9)
corners <- function(){
  # locator(1)은 마우스로 찍 점 하나의 위치 
  coos <- c(unlist(locator(1)), unlist(locator(1)))
  rect(coos[1], coos[2], coos[3], coos[4])
}
arrows(1,1,3,8, angle = 170)
corners()
```

```r
plot(0:10, 0:10, xlab = '', ylab = '', xaxt = 'n', yaxt = 'n', type = 'n')
rect(6,6,9,9)
click.arrows <- function(){
  coos <- c(unlist(locator(1)), unlist(locator(1)))
  arrows(coos[1],coos[2],coos[3],coos[4])
}
click.arrows()
```
