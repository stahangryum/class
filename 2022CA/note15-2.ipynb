```r
par(mfrow=c(1,1))

#set.seed(1)
n <- 150
y <- rnbinom(n, mu=1.5, size=1)
bks <- -0.5:(max(y) + 0.5)
hist(y, bks, main="")

m <- mean(y)
s <- mean(y)^2/(var(y)-mean(y))

xs <- 0:max(y)
ys <- dnbinom(xs, size=s, mu=m)
lines(xs, ys*n)
```
