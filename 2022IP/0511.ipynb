## 문제 3

아래 조건을 만족하는 클래스를 구현하라.

`-` 클래스 내에는 변수 a가 있다. 변수 a의 초기 값은 0이다.

`-` 클래스 내에는 show()라는 메소드가 있다. show()는 a의 값을 출력하는 기능을 한다.

`-` 클래스 내에는 up()라는 메소드가 있다. up()는 a의 값을 1 증가시키는 기능을 한다.

`-` 클래스 내에는 down()라는 메소드가 있다. down()는 a의 값을 1 감소시키는 기능을 한다.

```python
class Klass3:
    a = 0
    def show(self):
        print(self.a)
        
    def up(self):
        self.a = self.a + 1
        
    def down(self):
        self.a = self.a - 1
        
```

```python
ex3 = Klass3()
```

```python
ex3.show()
```

```python
ex3.up()
```

```python
ex3.show()
```

```python
ex3.down()
ex3.show()
```

# 문제 4

아래 조건을 만족하는 클래스를 구현하라.

`-` 클래스 내에는 변수 url이 있다. url의 초기값은 'https://github.com/guebin/IP2022/blob/master/_notebooks/2022-05-07-stop1.jpeg?raw=true'이다.

`-` 클래스 내에는 show()라는 메소드가 있고 아래와 같은 기능을 한다.

- 기능1: url의 그림을 출력

- 기능2: '당신은 이 그림을 $n$번 봤습니다.'를 출력. 여기에서 $n$은 그림을 본 횟수


```python
class Klass4:
    url = url1
    n = 0
    
    def show(self):
        self.n = self.n + 1
        display(Image.open(Image.io.BytesIO(requests.get(self.url).content)))
        print('당신은 이 그림을 {n}번 봤습니다.'.format(n = self.n))
```

```python
ex4 = Klass4()
```

```python
ex4.show()
```

# 숙제

아래 조건을 만족하는 클래스를 구현하라.

`-` 가위, 바위, 보 중 하나를 골라서 내는 메소드를 정의하라.

```python
import numpy as np
```

```python
np.random.choice(['가위', '바위', '보'])
```

```python
class rock_scissors_paper:
    def play(self):
        print(np.random.choice(['가위', '바위', '보']))
```

```python
쿵짝짝 = rock_scissors_paper()
```

```python
쿵짝짝.play()
```
