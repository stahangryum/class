# 숙제 

아래 조건을 만족하는 클래스를 구현하라.

`-` ['가위', '바위']와 같은 리스트를 입력으로 받아 인스턴스를 생성한다.

`-` 위의 리스트에서 하나의 값을 뽑는 메소드 `f`를 가지고 있다.

`-` 사용 예시

```python
a = Klass(['가위', '바위'])
a.f() # '가위'가 1/2, 바위가 1/2의 확률로 출력

b = Klass(['가위', '바위', '보'])
b.f() # '가위', '바위', '보'가 1/3의 확률로 출력
```

````python
import numpy as np
````

````python
class rock_scissors_paper:
    def __init__(self, weapons):
        self.weapons = weapons
    
    def f(self):
        print(np.random.choice(self.weapons, size = 1))
````

````python
a = rock_scissors_paper(['가위', '바위'])
````

````python
a.f()
````

````python
b = rock_scissors_paper(['가위', '바위', '보'])
````

````python
b.f()
````
